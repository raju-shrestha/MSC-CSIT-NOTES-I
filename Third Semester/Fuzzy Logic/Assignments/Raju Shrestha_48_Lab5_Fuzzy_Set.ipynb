Implement fuzzy membership functions as discussed in the class.
Your program should contain functions for R-function, L-function, triangular-function, and trapezoidal function. 
The program should ask required arguments for above functions and should generate fuzzy sets.

```python
import numpy as np
import matplotlib.pyplot as plt

# R-function
def r_function(x, a, b):
    if x <= a:
        return 0
    elif a < x < b:
        return (x - a) / (b - a)
    else:
        return 1
        
# L-function
def l_function(x, a, b):
    if x <= a:
        return 1
    elif a < x < b:
        return (b - x) / (b - a)
    else:
        return 0
        
# Triangular function
def triangular_function(x, a, b, c):
    if x <= a or x >= c:
        return 0
    elif a < x <= b:
        return (x - a) / (b - a)
    elif b < x < c:
        return (c - x) / (c - b)
    else:
        return 0

# Trapezoidal 
def trapezoidal_function(x, a, b, c, d):
    if x <= a or x >= d:
        return 0
    elif a < x <= b:
        return (x - a) / (b - a)
    elif b < x <= c:
        return 1
    elif c < x < d:
        return (d - x) / (d - c)
    else:
        return 0

# Function to generate a fuzzy set based on the user input
def generate_fuzzy_set(function, args):
    # Range for universe of discourse
    x = np.arange(10, 80, 10) # (start, end, step)
    y = [function(val, *args) for val in x]
    return x, y

# Main function to take user input   
def main():
    while True:
        print("Fuzzy Membership Functions")
        print("1. R-function")
        print("2. L-function")
        print("3. Triangular Function")
        print("4. Trapezoidal Function")
        print("5. Exit")
        choice = int(input("Enter your choice: "))
        
        if choice == 1:
            a = float(input("Enter the value of a: "))
            b = float(input("Enter the value of b: "))
            x, y = generate_fuzzy_set(r_function, [a, b])
            plt.plot(x, y)
            plt.title("R-function")
            
        elif choice == 2:
            a = float(input("Enter the value of a: "))
            b = float(input("Enter the value of b: "))
            x, y = generate_fuzzy_set(l_function, [a, b])
            plt.plot(x, y)
            plt.title("L-function")
        
        elif choice == 3:
            a = float(input("Enter the value of a: "))
            b = float(input("Enter the value of b: "))
            c = float(input("Enter the value of c: "))
            x, y = generate_fuzzy_set(triangular_function, [a, b, c])
            plt.plot(x, y)
            plt.title("Triangular function")
        
        elif choice == 4:
            a = float(input("Enter the value of a: "))
            b = float(input("Enter the value of b: "))
            c = float(input("Enter the value of c: "))
            d = float(input("Enter the value of d: "))
            x, y = generate_fuzzy_set(trapezoidal_function, [a, b, c,d])
            plt.plot(x, y)
            plt.title("Trapezoidal function")
        
        elif choice == 5:
            break
        
        else:
            print("Invalid choice. Please enter a number between 1 and 5.")

       
        plt.xlabel("x")
        plt.ylabel("Membership degree")
        plt.grid(True)
        plt.show()

if __name__ == "__main__":
    main()
```
